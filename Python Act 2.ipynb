```python
"Problem #2"
#Create empty list
initial_list = []

#get range from user input
n = int(input("Enter number of elements : "))

for a in range(0, n):
    b = int(input())#get values of list from user input
    initial_list.insert(a,b)#insert values on the list 

#printing of the first layer
print(initial_list)

#Modification of values in the list
for j in range(0,len(initial_list)-1):
  for i in range(0,len(initial_list)-1):
    x= initial_list[i]+initial_list[i+1]#making "x" the addition
    del initial_list[i]#deleting original value
    initial_list.insert(i,x)#adding x as new value
  del initial_list[len(initial_list)-1]#delete last value of the list
  print(initial_list)#print next layer
```
